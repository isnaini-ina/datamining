# Decision Tree

```python
import numpy as np
import pandas as pd
from sklearn.metrics import accuracy_score
from sklearn import tree
from matplotlib import pyplot as plt
```

```python
url = "https://gist.githubusercontent.com/netj/8836201/raw/6f9306ad21398ea43cba4f7d537619d0e07d5ae3/iris.csv"
data = pd.read_csv(url)
data
```

```python
# Variabel independen
X = data.drop(columns=["variety"])
#Cek apakah variabel target sudah dihapus
X.head()
```

```python
# Variabel dependen
Y = data["variety"]
Y.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 123)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(x_train)

x_train_scaled = scaler.transform(x_train)
x_test_scaled = scaler.transform(x_test)
```

```python
clf = tree.DecisionTreeClassifier(criterion="entropy")
clf = clf.fit(x_train_scaled, y_train)
```

```python
fig = plt.figure(figsize=(20,20))
_ = tree.plot_tree(clf, feature_names=list(data.columns.values)[:4], class_names=list(data.columns.values)[4] ,filled=True)
```

```python
print("Data Train : ", clf.score(x_train_scaled, y_train))
print("Data Test : ", clf.score(x_test_scaled, y_test))
```
