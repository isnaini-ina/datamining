```python
import pandas as pd
import numpy as np
import math
```

```python
dataset_url="https://raw.githubusercontent.com/isnaini-ina/datamining/main/Belgium.csv"
p=pd.read_csv(dataset_url, index_col=0)
```

```python
p
```

```python
def diss(x,y):
  a = 4
  b = 0
  for col in ["Country", "Confirmed Cases", "Deaths",	"New Cases",	"Country Population"]:
    if p[col][x] == p[col][y]:
      b += 1
    return (a-b)/a
```

```python
print(diss(1,1))
```

```python
p.describe()
```

```python
#normalisasi kb
std_kb = 1.505478e+06
mean_kb = 1.472642e+06
for col1 in ["Confirmed Cases"]:
  p1 = p[col1]
  hasil1 = (p1-mean_kb)/std_kb
  print(hasil1)

#normalisasi kt
std_kt = 9955.316835
mean_kt = 20381.694969
for col2 in ["Deaths"]:
  p2 = p[col2]
  hasil2 = (p2-mean_kt)/std_kt
  print(hasil2)
```

```python
#menghitung jarak data numerik dengan cara euclidean
def num_p(a,b,c):
  result = 0
  for col in [hasil1, hasil2]:
    result = math.pow(abs(col[a] -col[b]), c)
  return result**(1/c)
```

```python
print(num_p(1,2,2))
print(num_p(1,3,2))
print(num_p(1,4,2))
```
